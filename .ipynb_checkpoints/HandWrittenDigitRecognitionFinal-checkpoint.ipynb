```python
import numpy as np
from keras.datasets import mnist
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
(x_train_inp , y_train_inp ), (x_test_inp , y_test_inp)  = mnist.load_data()
x_train_inp = x_train_inp/255.0
converted_ans = []
```

```python
for i in range(len(y_train_inp)):
        ans = np.zeros(10)
        ans[y_train_inp[i]] = 1
        converted_ans.append(ans)
```

```python
size_of_input_layer = 784
size_of_hidden_layer = 100
size_of_output_layer = 10
```

```python
def initialize_weights(size_of_input_layer,number_of_neurons):
    weights_layer = np.random.randn(size_of_input_layer , number_of_neurons) 
    weights_layer = np.transpose(weights_layer)
    return weights_layer
```

```python
weights_ip = initialize_weights(size_of_input_layer,size_of_hidden_layer)
weights_hl = initialize_weights(size_of_hidden_layer,size_of_hidden_layer)
weights_op = initialize_weights(size_of_hidden_layer,size_of_output_layer)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_train_inp,converted_ans,test_size = 0.35,shuffle  = True)
```

```python
def activation_sigmoid(x):
    return 1/(1 + np.exp(-x))


def sigmoid_der(x):
    return (1 - x ) * x


def loss_function(real, predict):
    sum = np.sum((real - predict) ** 2)
    return sum



def predict_final(final_out):
    max = -2333
    index = -1
    for i in range(0,10):
        if (final_out[i] > max):
            max = final_out[i]
            index = i
    return index
```

```python
def predict_output(input_arr,weights_ip,weights_hl,weights_op):
    hl_output = activation_sigmoid(np.dot(weights_ip,input_arr))
    hl_output2 = activation_sigmoid(np.dot(weights_hl,hl_output))
    final_output = activation_sigmoid(np.dot(weights_op,hl_output2))
    if(final_output.sum() != 0.0 and final_output.min() != 0):
        final_output = final_output/final_output.min()
        final_output = final_output/final_output.sum()   
    return hl_output,hl_output2,final_output 
```

```python
def backpropagation(weights_ip,weights_hl,weights_op,input_data , hl_output,hl_output2,final_output,expected_output,learning_rate):
    change_op = 2 * (final_output - expected_output) * sigmoid_der(final_output)  # row wise multiply by *
#     print(2 * (final_output - expected_output))
    weight_op_update = learning_rate * np.outer(change_op, hl_output2)    # every number multiplied by every other number in other array
    weights_op -= weight_op_update
    change_hl = np.dot(change_op, weights_op) * sigmoid_der(hl_output2)
    weight_hl_update = learning_rate * np.outer(change_hl,hl_output)
    weights_hl -= weight_hl_update
    change_ip = np.dot(change_hl, weights_hl) * sigmoid_der(hl_output)
    weight_ip_update = learning_rate * np.outer(change_ip, input_data)
    weights_ip -= weight_ip_update
```

```python
epochs = 50
count = 0
total = 0
learning_r = 0.01
```

```python
for i in range(epochs):
    loss_sum = 0
    for j in range(len(x_train)):
        input_arr = np.reshape(x_train[j],784)
        hl_output ,hl_output2, final_output = predict_output(input_arr,weights_ip,weights_hl,weights_op)
        loss_sum += loss_function(final_output,y_train[j])
        if(predict_final(final_output) == predict_final(y_train[j])):
            count = count + 1
        backpropagation(weights_ip,weights_hl,weights_op,input_arr,hl_output,hl_output2,final_output,y_train[j],learning_r)
        total = total + 1
    print("Loss : ",loss_sum/len(x_train))
accuracy = (count / total) * 100.0 
accuracy
```
